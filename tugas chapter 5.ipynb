```python
import pandas as pd # Mengimpor library pandas untuk manipulasi dan analisis data tabular
from scipy.stats.mstats import gmean, hmean # Mengimpor fungsi rata-rata geometri dan harmonik dari library scipy
import matplotlib.pyplot as plt # Mengimpor modul pyplot dari matplotlib untuk visualisasi grafik

my_dataset = pd.read_excel('Smith_glass_post_NYT_data.xlsx', sheet_name='Supp_traces') # Membaca data dari file Excel spesifik pada sheet 'Supp_traces' ke dalam DataFrame

a_mean = my_dataset.Zr.mean() # Menghitung nilai rata-rata aritmatika dari kolom populasi 'Zr'
g_mean = gmean(my_dataset['Zr']) # Menghitung nilai rata-rata geometri dari kolom populasi 'Zr'
h_mean = hmean(my_dataset['Zr']) # Menghitung nilai rata-rata harmonik dari kolom populasi 'Zr'

print ('-------') # Mencetak pembatas visual pada output
print ('arithmetic mean') # Mencetak label teks untuk nilai rata-rata aritmatika
print ("{0:.1f} [ppm]".format(a_mean)) # Mencetak nilai rata-rata aritmatika dengan presisi satu angka desimal beserta satuannya
print ('-------') # Mencetak pembatas visual pada output

print ('geometric mean') # Mencetak label teks untuk nilai rata-rata geometri
print ("{0:.1f} [ppm]".format(g_mean)) # Mencetak nilai rata-rata geometri dengan presisi satu angka desimal beserta satuannya
print ('-------') # Mencetak pembatas visual pada output

print ('harmonic mean') # Mencetak label teks untuk nilai rata-rata harmonik
print ("{0:.1f} [ppm]".format(h_mean)) # Mencetak nilai rata-rata harmonik dengan presisi satu angka desimal beserta satuannya
print ('-------') # Mencetak pembatas visual pada output

fig, ax = plt.subplots() # Menginisialisasi objek figure dan axes untuk kanvas grafik
ax.hist(my_dataset.Zr, bins='auto', density=True, edgecolor='k', label='Measurements Hist', alpha=0.8)  # Membuat plot histogram dari data Zr dengan lebar kelas (bins) otomatis
ax.axvline(a_mean, color='purple', label='Arithmetic mean', linewidth=2) # Menambahkan garis vertikal penanda untuk posisi rata-rata aritmatika
ax.axvline(g_mean, color='orange',  label='Geometric mean', linewidth=2) # Menambahkan garis vertikal penanda untuk posisi rata-rata geometri
ax.axvline(h_mean, color='green',  label='Harmonic mean', linewidth=2) # Menambahkan garis vertikal penanda untuk posisi rata-rata harmonik
ax.set_xlabel('Zr [ppm]') # Menetapkan label teks untuk sumbu X
ax.set_ylabel('Probability density') # Menetapkan label teks untuk sumbu Y
ax.legend() # Memunculkan legenda grafik berdasarkan label yang didefinisikan
plt.show() # Merender dan menampilkan keseluruhan grafik ke layar
```

```python
import pandas as pd # Mengimpor modul pandas untuk pengolahan dataset
import numpy as np # Mengimpor modul numpy untuk operasi matematika tingkat lanjut
import matplotlib.pyplot as plt # Mengimpor modul matplotlib untuk visualisasi histogram

my_dataset = pd.read_excel('Smith_glass_post_NYT_data.xlsx', sheet_name='Supp_traces') # Mengakses dan membaca file sumber data Excel

hist, bin_edges = np.histogram(my_dataset['Zr'], bins= 20, density=True) # Menghitung frekuensi (hist) dan batas interval (bin_edges) dengan membagi data menjadi 20 kelompok
modal_value = (bin_edges[hist.argmax()] +  bin_edges[hist.argmax()+1])/2 # Menentukan nilai modus dengan mencari nilai tengah dari interval kelas yang memiliki frekuensi tertinggi

print ('modal value: {0:.0f} [ppm]'.format(modal_value)) # Mencetak nilai modus hasil perhitungan tanpa angka desimal

fig, ax = plt.subplots() # Membuat subplots baru sebagai wadah visualisasi
ax.hist(my_dataset.Zr, bins=20, density=True, edgecolor='k', label="Measurements Hist", alpha=0.8) # Membuat plot diagram batang (histogram) dari kumpulan data Zr
ax.axvline(modal_value, color="orange", label="Modal value", linewidth=2) # Menggambar garis tegak lurus sebagai penanda posisi modus pada sumbu X
ax.set_xlabel('Zr [ppm]') # Mendefinisikan label sumbu absis (X)
ax.set_ylabel('Probability density') # Mendefinisikan label sumbu ordinat (Y)
ax.legend() # Menampilkan boks legenda di dalam area grafik
plt.show() # Menginstruksikan program untuk merender grafik ke antarmuka pengguna
```

```python
import pandas as pd # Mengimpor library pandas untuk analisis data
import matplotlib.pyplot as plt # Mengimpor library matplotlib untuk plotting grafik

my_dataset = pd.read_excel('Smith_glass_post_NYT_data.xlsx', sheet_name='Supp_traces') # Membaca dan menyimpan data sheet 'Supp_traces' ke variabel DataFrame

variance =  my_dataset['Zr'].var() # Menghitung nilai varians (keragaman) dari kolom populasi Zr
stddev =  my_dataset['Zr'].std() # Menghitung nilai standar deviasi (simpangan baku) dari kolom populasi Zr

print ('-------') # Mencetak garis pemisah
print ('Variance') # Menampilkan teks keterangan untuk nilai varians
print("{0:.0f} [square ppm]".format(variance)) # Mencetak nilai varians hasil perhitungan
print ('-------') # Mencetak garis pemisah
print ('Standard Deviation') # Menampilkan teks keterangan untuk nilai standar deviasi
print("{0:.0f} [ppm]".format(stddev)) # Mencetak nilai standar deviasi hasil perhitungan
print ('-------') # Mencetak garis pemisah

fig, ax = plt.subplots() # Mempersiapkan kanvas dan sumbu untuk pembuatan plot
ax.hist(my_dataset.Zr, bins= 20, density = True, edgecolor='k', label='Measurements Hist') # Membuat histogram dengan 20 interval kelas data
ax.axvline(my_dataset['Zr'].mean()-stddev, color='purple', label=r'mean - 1$\sigma$', linewidth=2) # Membuat garis batas bawah standar deviasi (Mean dikurangi 1 Sigma)
ax.axvline(my_dataset['Zr'].mean()+stddev, color='green', label=r'mean + 1$\sigma$', linewidth=2) # Membuat garis batas atas standar deviasi (Mean ditambah 1 Sigma)
ax.axvspan(my_dataset['Zr'].mean()-stddev, my_dataset['Zr'].mean()+stddev, alpha=0.1, color='orange', label=r'mean $\pm$ 1$\sigma$') # Mewarnai area jangkauan satu standar deviasi dari rata-rata (mencakup ~68% data)
ax.set_xlabel('Zr [ppm]') # Menetapkan label sumbu X
ax.set_ylabel('Probability density') # Menetapkan label sumbu Y
ax.legend() # Mengaktifkan tampilan legenda grafik
plt.show() # Mengeksekusi dan menampilkan grafik secara penuh
```

```python
import pandas as pd # Mengimpor modul pandas untuk manipulasi data tabel
import numpy as np # Mengimpor modul numpy untuk komputasi numerik dan statistik
import matplotlib.pyplot as plt # Mengimpor modul matplotlib untuk visualisasi data

my_dataset = pd.read_excel('Smith_glass_post_NYT_data.xlsx', sheet_name='Supp_traces') # Mengimpor data dari file Excel spesifik ke dalam bentuk DataFrame

q1 = np.percentile(my_dataset.Zr, 25, interpolation = 'midpoint') # Menghitung kuartil pertama (Q1) atau persentil ke-25 dari distribusi data Zr
q3 = np.percentile(my_dataset.Zr, 75, interpolation = 'midpoint') # Menghitung kuartil ketiga (Q3) atau persentil ke-75 dari distribusi data Zr
  
iqr = q3 - q1 # Menghitung Interquartile Range (IQR) dengan mencari selisih antara Q3 dan Q1

print ('-------') # Mencetak pembatas visual
print ('Interquaritle range (IQR): {0:.0f} [ppm]'.format(iqr)) # Menampilkan hasil perhitungan IQR tanpa angka desimal
print ('-------') # Mencetak pembatas visual

fig, ax = plt.subplots() # Membuat figure dan axes untuk wadah plot
ax.hist(my_dataset.Zr, bins='auto', density=True, edgecolor='k', label='Measurements Hist') # Membuat histogram distribusi frekuensi data Zr
ax.axvline(q1, color='purple', label='Q1', linewidth=2) # Menambahkan garis vertikal penanda batas Kuartil 1
ax.axvline(q3, color='green', label='Q3', linewidth=2) # Menambahkan garis vertikal penanda batas Kuartil 3
ax.axvspan(q1, q3, alpha=0.1, color='orange', label='Interquaritle range (IQR)') # Membuat blok warna area untuk merepresentasikan rentang IQR (50% data tengah)
ax.set_xlabel('Zr [ppm]') # Menambahkan keterangan label pada sumbu X
ax.set_ylabel('Probability density') # Menambahkan keterangan label pada sumbu Y
ax.legend() # Menampilkan legenda dari masing-masing plot penanda
plt.show() # Menampilkan keseluruhan plot ke jendela keluaran
```

```python
import pandas as pd # Memanggil library pandas untuk membaca file Excel
import numpy as np # Memanggil library numpy untuk analisis fungsi matematis tingkat lanjut
import matplotlib.pyplot as plt # Memanggil library matplotlib untuk memvisualisasikan data

my_dataset = pd.read_excel('Smith_glass_post_NYT_data.xlsx', sheet_name='Supp_traces') # Mengambil dataset dari Excel untuk dimasukkan ke struktur data DataFrame

a_mean = my_dataset.Zr.mean() # Menghitung parameter mean (rata-rata) aritmatika dari populasi data Zr

median = my_dataset.Zr.median() # Menghitung parameter median (nilai tengah) dari distribusi data Zr

hist, bin_edges = np.histogram(my_dataset['Zr'], bins=20, density=True) # Menghitung frekuensi data (hist) dan batas interval (bin_edges) untuk identifikasi kelas modus
modal_value = (bin_edges[hist.argmax()] +  bin_edges[hist.argmax()+1])/2 # Mengkalkulasi estimasi nilai modus dengan mengambil titik tengah dari interval kelas frekuensi tertinggi

fig, ax = plt.subplots() # Membuat subplots untuk meletakkan grafik histogram gabungan
ax.hist(my_dataset.Zr, bins=20, density=True, edgecolor='k', label="Measurements Hist") # Melakukan plotting distribusi data Zr ke dalam 20 kelompok bar
ax.axvline(modal_value, color='orange', label='Modal Value', linewidth=2) # Menambahkan penanda garis vertikal khusus untuk lokasi nilai Modus
ax.axvline(median, color='purple',  label='Median Value', linewidth=2) # Menambahkan penanda garis vertikal khusus untuk lokasi nilai Median
ax.axvline(a_mean, color='green',  label='Arithmetic mean', linewidth=2) # Menambahkan penanda garis vertikal khusus untuk lokasi nilai Mean
ax.set_xlabel('Zr [ppm]') # Memberikan keterangan teks representasi parameter pada sumbu horizontal
ax.set_ylabel('Probability density') # Memberikan keterangan teks representasi fungsi pada sumbu vertikal
ax.legend() # Memunculkan kotak legenda untuk membedakan ketiga ukuran pemusatan data
plt.show() # Menampilkan plot ke antarmuka IDE pengguna
```

```python
import pandas as pd # Mengimpor library pandas untuk penanganan data tabular
import numpy as np # Mengimpor numpy untuk pemrosesan array dan fungsi histogram
from scipy.stats import skew # Mengimpor fungsi skew dari modul scipy.stats untuk mengukur kemencengan kurva (skewness)

my_dataset = pd.read_excel('Smith_glass_post_NYT_data.xlsx', sheet_name='Supp_traces') # Memuat sumber data eksperimental dari file Excel

a_mean = my_dataset.Zr.mean() # Mengekstraksi nilai rata-rata sampel dari atribut Zr
median = my_dataset.Zr.median() # Mengekstraksi nilai tengah (median) dari atribut Zr
hist, bin_edges = np.histogram(my_dataset['Zr'], bins=20, density=True) # Mendistribusikan data ke dalam bins untuk memfasilitasi pencarian nilai maksimum lokal
modal_value = (bin_edges[hist.argmax()] + bin_edges[hist.argmax()+1])/2 # Mengestimasi nilai modus populasi dari representasi bin tertinggi
standard_deviation = my_dataset['Zr'].std() # Menghitung simpangan baku (standar deviasi) sampel

a1 = (a_mean - modal_value) / standard_deviation # Mengaplikasikan rumus koefisien kemencengan Pearson pertama (menggunakan referensi modus)
a2 = 3 * (a_mean - median) / standard_deviation # Mengaplikasikan rumus momen kemencengan Pearson kedua (menggunakan referensi median)
g1 = skew(my_dataset['Zr']) # Menghitung koefisien kemencengan Fisher-Pearson menggunakan algoritma standar dari library Scipy

print ('-------') # Menampilkan pemisah baris pada standard output
print ("Pearson's first coefficient of skewness: {:.2f}".format(a1)) # Mencetak output perhitungan Skewness Pearson pertama dengan format presisi 2 desimal
print ("Pearson's 2nd moment of skewness: {:.2f}".format(a2)) # Mencetak output perhitungan Skewness Pearson kedua dengan format presisi 2 desimal
print ("Fisher-Pearson's coefficient of skewness: {:.2f}".format(g1)) # Mencetak output perhitungan Skewness Fisher-Pearson dengan format presisi 2 desimal
print ('-------') # Menampilkan pemisah baris untuk penutup output
```

```python
import pandas as pd # Memanggil library utama pandas untuk manajemen dataset
import matplotlib.pyplot as plt # Memanggil pustaka fungsional matplotlib.pyplot untuk inisialisasi dan kustomisasi plot visual

my_dataset = pd.read_excel('Smith_glass_post_NYT_data.xlsx', sheet_name='Supp_traces') # Memasukkan struktur data excel ke dalam memori variabel aplikasi

fig, ax = plt.subplots() # Membuat instance dasar figure dan axes untuk menyusun lapisan plot
my_flierprops = dict(markerfacecolor="#f9d006", markeredgecolor='#24305e', marker='*') # Mengkonfigurasi parameter visual (kode warna dan metrik bentuk) untuk representasi data pencilan (outliers)
my_medianprops = dict(color="#f61509", linewidth = 2) # Mengkonfigurasi parameter ketebalan dan pewarnaan garis median di dalam kuartil
my_boxprops = dict(facecolor="#08fa10", edgecolor='#24305e') # Mengkonfigurasi parameter visual (warna isi dan warna batas tepi) untuk merepresentasikan area rentang antar kuartil (IQR)
ax.boxplot(my_dataset.Zr, patch_artist = True, notch=True, flierprops=my_flierprops, medianprops=my_medianprops, boxprops=my_boxprops) # Membangun objek boxplot parametrik dengan mengaplikasikan argumen properti visual yang telah didefinisikan sebelumnya
ax.set_ylabel('Zr [ppm]') # Menetapkan deskripsi representasi data tunggal pada sumbu ordinat (Y)
ax.set_xticks([1]) # Menginisialisasi metrik lokasi titik penanda (tick) tunggal pada sumbu absis (X)
ax.set_xticklabels(['all Epochs']) # Memetakan penamaan label secara eksplisit untuk penanda posisi sumbu absis
plt.show() # Melakukan kompilasi grafis dan merender objek plot untuk divisualisasikan
```
